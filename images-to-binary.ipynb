```python
import numpy as np
import cv2
import glob
import os
```

```python
targetFolder = "gtFine/train/delamination/"
output_file = "cityscapes_train_list"
targetExtension = '.png'
```

```python
targetList = os.listdir(targetFolder)
targetList.sort()
```

```python
thresh = 10
CLASS = 11
for path in targetList:
    img = np.array(cv2.imread(targetFolder + path, 0))
    img[img < thresh] = 0
    img[img >= thresh] = CLASS
    cv2.imwrite(targetFolder+ "/out/" + path, img)
    print(np.unique(img))
```
